```python
import pandas as pd
import numpy as np
```

## Read 2019 Utah PUMS data 

```python
#2019 1-YEAR data
pums_data_1 = pd.read_csv (r"E:\GitHub\TDM-Value-Of-Time-Calculation\censusdata\1-Year 2019 PUMS\psam_h49.csv")
pums_data_1['TIMEFRAME'] = "1-Year ACS"
display(pums_data_1)

#2019 5-YEAR data
pums_data_5 = pd.read_csv (r"E:\GitHub\TDM-Value-Of-Time-Calculation\censusdata\5-Year 2019 PUMS\psam_h49.csv")
pums_data_5['TIMEFRAME'] = "5-Year ACS"
display(pums_data_5)
```

```python
display(len(pums_data_1))
display(len(pums_data_5))

#concat
pums_data = pd.concat([pums_data_1,pums_data_5])
display(len(pums_data))
```

```python
pums_data = pums_data.reset_index()
pums_data
```

## Drop GQ

```python
pums_data = pums_data.drop(pums_data[pums_data.TYPE > 1].index)
len(pums_data)
```

## Drop Empty Households

```python
pums_data = pums_data.drop(pums_data[pums_data.NP == 0].index)
len(pums_data)
display(len(pums_data[pums_data['TIMEFRAME']=='1-Year']))
display(len(pums_data[pums_data['TIMEFRAME']=='5-Year']))
```

## Drop PUMAs outside of WFRC

```python
pums_data = pums_data.drop(pums_data[pums_data.PUMA == 13001].index)
pums_data = pums_data.drop(pums_data[pums_data.PUMA == 21001].index)
pums_data = pums_data.drop(pums_data[pums_data.PUMA == 49001].index)
pums_data = pums_data.drop(pums_data[pums_data.PUMA == 49002].index)
pums_data = pums_data.drop(pums_data[pums_data.PUMA == 49003].index)
pums_data = pums_data.drop(pums_data[pums_data.PUMA == 49004].index)
len(pums_data)
```

## Adjust income to 2019 Dollars

```python
# You have to apply the adjustment factor to household income to scale to one year (2019)
# ADJINC Character 7
# Adjustment factor for income and earnings dollar amounts (6 implied
# decimal places)
# 1080470 .2015 factor (1.001264 * 1.07910576)
# 1073449 .2016 factor (1.007588 * 1.06536503)
# 1054606 .2017 factor (1.011189 * 1.04293629)
# 1031452 .2018 factor (1.013097 * 1.01811790)
# 1010145 .2019 factor (1.010145 * 1.00000000)
# C. Note on Income and Earnings Inflation Factor (ADJINC)
# Divide ADJINC by 1,000,000 to obtain the inflation adjustment factor and multiply it to
# the PUMS variable value to adjust it to 2019 dollars. Variables requiring ADJINC on the
# Housing Unit file are FINCP and HINCP. Variables requiring ADJINC on the Person
# files are: INTP, OIP, PAP, PERNP, PINCP, RETP, SEMP, SSIP, SSP, and WAGP.
```

```python
pums_data['HINCP_ADJ'] = pums_data['HINCP'] * pums_data['ADJINC'] / 1000000
pums_data['HINCP_ADJ']
```

## Plot capped income distribution

```python
pums_data['HINCP_CAPPED'] = pums_data['HINCP_ADJ'].clip(lower=0, upper=500000)

display(pums_data[pums_data['TIMEFRAME']=='1-Year ACS'].hist(column="HINCP_CAPPED", bins=100))
display(pums_data[pums_data['TIMEFRAME']=='5-Year ACS'].hist(column="HINCP_CAPPED", bins=100))
```

## Expand data by person and expansion weight (could also expand by average trips per hh or person)

```python
pums_exp_data = pums_data.loc[pums_data.index.repeat(pums_data.NP)]
pums_exp_data = pums_exp_data.reset_index(drop=True)
pums_exp_data = pums_exp_data.loc[pums_exp_data.index.repeat(pums_exp_data.WGTP)]
pums_exp_data = pums_exp_data.reset_index(drop=True)
len(pums_exp_data)
```

```python
## Draw time sensitivity from distribution
```

```python
elements = len(pums_exp_data)
TIME_FACTOR_WORK= np.random.lognormal(0.0, 0.7,elements)
pums_exp_data["TIME_FACTOR_WORK"] = TIME_FACTOR_WORK
pums_exp_data["TIME_FACTOR_WORK"] = pums_exp_data["TIME_FACTOR_WORK"].clip(0.1,10)
display(pums_exp_data[pums_exp_data['TIMEFRAME']=='1-Year ACS'].hist(column="TIME_FACTOR_WORK", bins=100))
display(pums_exp_data[pums_exp_data['TIMEFRAME']=='5-Year ACS'].hist(column="TIME_FACTOR_WORK", bins=100))
```

```python
TIME_FACTOR_NWRK= np.random.lognormal(0.0, 0.6,elements)
pums_exp_data["TIME_FACTOR_NWRK"] = TIME_FACTOR_NWRK
pums_exp_data["TIME_FACTOR_NWRK"] = pums_exp_data["TIME_FACTOR_NWRK"].clip(0.1,10)
pums_exp_data[pums_exp_data['TIMEFRAME']=='1-Year ACS'].hist(column="TIME_FACTOR_NWRK", bins=100)
pums_exp_data[pums_exp_data['TIMEFRAME']=='5-Year ACS'].hist(column="TIME_FACTOR_NWRK", bins=100)
```

## Calculate cost coefficients and value of time by occupancy

```python
C_IVT_WORK = -0.030
C_IVT_NWRK = -0.020
C_COST_WORK= -1.250 #Note this is scaled by household income raised to a fractional power
C_COST_NWRK= -0.524 #Note this is scaled by household income raised to a fractional power

pums_exp_data['WORK_SOV_COST_COEFF'] = C_COST_WORK * 1.0/pums_exp_data['HINCP_CAPPED'].clip(lower=1000).pow(0.6)
pums_exp_data['NWRK_SOV_COST_COEFF'] = C_COST_NWRK * 1.0/pums_exp_data['HINCP_CAPPED'].clip(lower=1000).pow(0.5)

pums_exp_data['WORK_S2_COST_COEFF'] = pums_exp_data['WORK_SOV_COST_COEFF']*0.62  #1/.62 = 1.6129 - Use when calculating path cost
pums_exp_data['WORK_S3_COST_COEFF'] = pums_exp_data['WORK_SOV_COST_COEFF']*0.42  #1/.42 = 2.3810 - Use when calculating path cost
pums_exp_data['NWRK_S2_COST_COEFF'] = pums_exp_data['NWRK_SOV_COST_COEFF']*0.62
pums_exp_data['NWRK_S3_COST_COEFF'] = pums_exp_data['NWRK_SOV_COST_COEFF']*0.42

pums_exp_data['WORK_SOV_VOT'] = ((C_IVT_WORK * pums_exp_data["TIME_FACTOR_WORK"])/pums_exp_data['WORK_SOV_COST_COEFF']) * 0.6
pums_exp_data['WORK_S2_VOT' ] = ((C_IVT_WORK * pums_exp_data["TIME_FACTOR_WORK"])/pums_exp_data['WORK_S2_COST_COEFF' ]) * 0.6
pums_exp_data['WORK_S3_VOT' ] = ((C_IVT_WORK * pums_exp_data["TIME_FACTOR_WORK"])/pums_exp_data['WORK_S3_COST_COEFF' ]) * 0.6
pums_exp_data['NWRK_SOV_VOT'] = ((C_IVT_NWRK * pums_exp_data["TIME_FACTOR_NWRK"])/pums_exp_data['NWRK_SOV_COST_COEFF']) * 0.6
pums_exp_data['NWRK_S2_VOT' ] = ((C_IVT_NWRK * pums_exp_data["TIME_FACTOR_NWRK"])/pums_exp_data['NWRK_S2_COST_COEFF' ]) * 0.6
pums_exp_data['NWRK_S3_VOT' ] = ((C_IVT_NWRK * pums_exp_data["TIME_FACTOR_NWRK"])/pums_exp_data['NWRK_S3_COST_COEFF' ]) * 0.6

pums_exp_data[pums_exp_data['TIMEFRAME']=='1-Year ACS'].hist(column="WORK_SOV_COST_COEFF", bins=100)
pums_exp_data[pums_exp_data['TIMEFRAME']=='5-Year ACS'].hist(column="WORK_SOV_COST_COEFF", bins=100)


#USE HHS TO CALCULATE TRIP RATES BY INCOME... SCALE AND RECALCULATE
```

```python
pd.options.display.float_format = '{:,.3f}'.format
dQuantileDivisions = [
    [      0.167,"1-VOT1 Mid" ,0],
    [      0.300,"2-VOT1 90th",1],
    [      0.317,"3-VOT1 95th",0],
    [      0.333,"4-VOT1 Max" ,0],
    [      0.500,"5-VOT2 Mid" ,0],
    [      0.633,"6-VOT2 90th",1],
    [      0.650,"7-VOT2 95th",0],
    [      0.667,"8-VOT2 Max" ,0],
    [      0.833,"9-VOT3 Mid" ,0],
    [      0.967,"A-VOT3 90th",1],
    [      0.983,"B-VOT3 95th",0],
    [      0.999,"C-VOT3 Max" ,0],
]
dfQuantileDivisions = pd.DataFrame(dQuantileDivisions, columns = ['QUANTILE','QUANTILEDESC',"USE"])
dfQuantileDivisions
```

```python
# Generate Quantiles for each dimension

dTimeBins = ['5-Year ACS']   #['1-Year ACS','5-Year ACS']
dPurpBins = ['WORK','NWRK']
dOccpBins = ['SOV','S2','S3']

dfQuantiles = pd.DataFrame()

for t in dTimeBins:
    for p in dPurpBins:
        for o in dOccpBins:
            _dfTemp = pums_exp_data[pums_exp_data['TIMEFRAME']==t][p + "_" + o + "_VOT"].quantile(q=dfQuantileDivisions['QUANTILE'].to_list()).to_frame()
            _dfTemp = _dfTemp.reset_index()
            _dfTemp = _dfTemp.rename(columns={p + "_" + o + "_VOT":"VOT","index":"QUANTILE"})
            _dfTemp['TIMEFRAME'] = t
            _dfTemp['PURPOSE'  ] = p
            _dfTemp['OCCUPANCY'] = o

            dfQuantiles = pd.concat([dfQuantiles,_dfTemp])

dfQuantilesWithQDesc = pd.DataFrame.merge(dfQuantiles,dfQuantileDivisions,on='QUANTILE')
display(dfQuantilesWithQDesc)
```

```python
pd.options.display.float_format = '{:,.2f}'.format

dfQuantiles_XT = dfQuantilesWithQDesc
dfQuantiles_XT.loc[(dfQuantiles_XT['OCCUPANCY']=='SOV'),'OCCUPANCY'] = 'DA'
dfQuantiles_XT = dfQuantiles_XT.pivot_table(index=['PURPOSE','OCCUPANCY','TIMEFRAME'],values='VOT',columns=('QUANTILEDESC'))
dfQuantiles_XT
```

```python
# convert to cents/minute
dfQuantiles_XT_centsperminute = dfQuantiles_XT / 0.6 # convert to cents per mile
dfQuantiles_XT_centsperminute
```

```python

dfQuantiles_XT_final = dfQuantilesWithQDesc[dfQuantilesWithQDesc['USE']==1]
dfQuantiles_XT_final.loc[(dfQuantiles_XT_final['OCCUPANCY']=='SOV'),'OCCUPANCY'] = 'DA'
dfQuantiles_XT_final = dfQuantiles_XT_final.pivot_table(index=['PURPOSE','OCCUPANCY','TIMEFRAME'],values='VOT',columns=('QUANTILEDESC'))
dfQuantiles_XT_final_centspermile = dfQuantiles_XT_final / 0.6 # convert to cents per mile
dfQuantiles_XT_final_centspermile
```
